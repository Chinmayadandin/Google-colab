```python
property_details = data[['id', 'start_date', 'end_date', 'created_on', 'lat', 'lon', 'l1', 'l2', 'l3', 'rooms', 'bedrooms', 'bathrooms', 'surface_total', 'surface_covered']]
property_price_details = data[['id', 'price', 'currency', 'price_period', 'property_type', 'operation_type']]
```

```python
import sqlite3

conn = sqlite3.connect('/content/drive/MyDrive/DE_SOLUTION_Chinmaya_Dandin/Colab Notebooks/housing.db')
```

```python
property_details.to_sql('property_details', conn, if_exists='replace', index=False)
property_price_details.to_sql('property_price_details', conn, if_exists='replace', index=False)
```

```python
#Retrieve properties that have a price greater than 1 million and are located in "Estados Unidos" (l1).

query = """
SELECT
    pd.id,
    pd.start_date,
    pd.end_date,
    pd.created_on,
    pd.lat,
    pd.lon,
    pd.l1,
    pd.l2,
    pd.l3,
    pd.rooms,
    pd.bedrooms,
    pd.bathrooms,
    pd.surface_total,
    pd.surface_covered,
    pp.price,
    pp.currency,
    pp.price_period,
    pp.property_type,
    pp.operation_type
FROM
    property_details pd
INNER JOIN
    property_price_details pp
ON
    pd.id = pp.id
WHERE
    pd.l1 = 'Estados Unidos'
    AND pp.price > 1000000;

"""

cursor = conn.cursor()
cursor.execute(query)
results1 = cursor.fetchall()
for row in results1:
    print(row)
```

```python
# 2. Categorize properties based on their surface area as 'Small' if it's less than 50 square meters, 'Medium' if it's
# between 50 and 100 square meters, and 'Large' if it's greater than 100 square meters:
query = """
SELECT
    *,
    CASE
        WHEN surface_total < 50 THEN 'Small'
        WHEN surface_total BETWEEN 50 AND 100 THEN 'Medium'
        WHEN surface_total > 100 THEN 'Large'
        ELSE 'Unknown'  -- Handle any other cases
    END AS surface_category
FROM
    property_details;


"""

cursor = conn.cursor()
cursor.execute(query)

results2 = cursor.fetchall()
for row in results2:
    print(row)
```

```python
# 3.List all properties (id) in the "Belgrano" neighborhood (l3) that have the same number of bedrooms and
# bathrooms as another property in the dataset:

query = """
SELECT DISTINCT pd1.id
FROM Property_Details pd1
JOIN Property_Details pd2 ON pd1.id <> pd2.id
WHERE pd1.l3 = 'Belgrano'
  AND pd1.bedrooms = pd2.bedrooms
  AND pd1.bathrooms = pd2.bathrooms;

"""
cursor=conn.cursor()
cursor.execute(query)

results3 = cursor.fetchall()
for row in results3:
    print(row)
```

```python
#4.Calculate the average price per square meter (price / surface_total) for each property type (property_type) in
#the "Belgrano" neighborhood (l3):
query = """
SELECT
    ppd.property_type,
    AVG(ppd.price / pd.surface_total) AS avg_price_per_sqm
FROM
    Property_Price_Details ppd
JOIN
    Property_Details pd ON ppd.id = pd.id
WHERE
    pd.l3 = 'Belgrano'
GROUP BY
    ppd.property_type;





"""

cursor = conn.cursor()
cursor.execute(query)

results4 = cursor.fetchall()
for row in results4:
    print(row)
```

```python
#5.Identify properties that have a higher price than the average price of properties with the same number of
# bedrooms and bathrooms.
query = """
SELECT pd.id, pd.start_date, pd.end_date, pd.created_on, pd.lat, pd.lon, pd.l1, pd.l2, pd.l3, pd.rooms, pd.bedrooms, pd.bathrooms,
       pd.surface_total, pd.surface_covered, pp.price, pp.currency, pp.price_period, pp.property_type, pp.operation_type
FROM Property_Details pd
JOIN Property_Price_Details pp ON pd.id = pp.id
WHERE pp.price > (
    SELECT AVG(price)
    FROM Property_Price_Details
    WHERE bedrooms = pd.bedrooms AND bathrooms = pd.bathrooms
);

"""
cursor=conn.cursor()
cursor.execute(query)

results5 = cursor.fetchall()
for row in results5:
    print(row)
```

```python
#6.Calculate the cumulative price for each property type, ordered by the creation date.
query = """


SELECT
    p.id AS property_id,
    pp.property_type,
    p.created_on,
    pp.price,
    pp.currency,
    pp.price_period,
    SUM(pp.price) OVER (PARTITION BY pp.property_type ORDER BY p.created_on) AS cumulative_price
FROM
    Property_Details p
JOIN
    Property_Price_Details pp
ON
    p.id = pp.id
ORDER BY
    pp.property_type,
    p.created_on;



"""
cursor = conn.cursor()
cursor.execute(query)

results6 = cursor.fetchall()
for row in results6:
    print(row)
```

```python
# 7. Identify the 10 locations (l3) with the highest total surface area (sum of surface_total) of properties listed for
# sale (operation_type = 'Venta'):
query = """
SELECT l3, SUM(surface_total) AS total_surface_area
FROM Property_Details pd
INNER JOIN Property_Price_Details ppd ON pd.id = ppd.id
WHERE ppd.operation_type = 'Venta'
GROUP BY l3
ORDER BY total_surface_area DESC
LIMIT 10;
"""

cursor = conn.cursor()
cursor.execute(query)

results7 = cursor.fetchall()
for row in results7:
    print(row)
```

```python
# 8.Find the top 5 most expensive properties (based on price) in the "Palermo" neighborhood (l3) that were listed
# in August 2020:
cursor.execute("""
SELECT pd.id, ppd.price
FROM property_details AS pd
JOIN property_price_details AS ppd ON pd.id = ppd.id
WHERE pd.l3 = 'Palermo' AND strftime('%m-%Y', pd.created_on) = '08-2020'
ORDER BY ppd.price DESC
LIMIT 5
""")



results8 = cursor.fetchall()
for row in results8:
    print(row)

#NO DATA FOUND
```

```python
# 9. Find the top 3 properties with the highest price per square meter (price divided by surface area) within each
# property type.
query = """
SELECT
    id,
    property_type,
    price_per_sq_meter
FROM (
    SELECT
        pd.id,
        pp.property_type,
        pp.price / pd.surface_total AS price_per_sq_meter,
        ROW_NUMBER() OVER (PARTITION BY pp.property_type ORDER BY pp.price / pd.surface_total DESC) AS row_num
    FROM
        Property_Details pd
    JOIN
        Property_Price_Details pp ON pd.id = pp.id
) AS ranked
WHERE
    row_num <= 3
  LIMIT 3;





"""

cursor = conn.cursor()
cursor.execute(query)

results9 = cursor.fetchall()
for row in results9:
    print(row)
```

```python
# Find the top 3 locations (l1, l2, l3) with the highest average price per square meter (price / surface_total) for
# properties listed for sale (operation_type = 'Venta') in the year 2020. Exclude locations with fewer than 10
# properties listed for sale in 2020 from the results.
query = """
SELECT
    l1,
    l2,
    l3,
    AVG(price / surface_total) AS avg_price_per_sqm
FROM
    Property_Details AS PD
JOIN
    Property_Price_Details AS PPD
ON
    PD.id = PPD.id
WHERE
    PPD.operation_type = 'Venta'
    AND PD.created_on BETWEEN '2020-01-01' AND '2020-12-31'
GROUP BY
    l1, l2, l3
HAVING
    COUNT(*) >= 10
ORDER BY
    avg_price_per_sqm DESC
LIMIT 3;

"""

cursor = conn.cursor()
cursor.execute(query)

results10 = cursor.fetchall()
for row in results10:
    print(row)

#NO DATA FOUND
```
